```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension("tabulator")
import hvplot.pandas
```

```python
df = pd.read_csv("owid-co2-data.csv")
df
```

```python
df.columns
```

```python
df[df['country'] == "World"]
```

```python
#sum the NA values in the dataframe
df.isna().sum()
```

```python
#Data pre-processing for dashboard purposes
#Fill NAs with 0s and create GDP per capital column
df = df.fillna(0)
df["gdp_per_capita"] = np.where(df['population'] != 0 , df['gdp']/ df['population'], 0)
df['gdp_per_capita']
```

CO2 emmissions over time by continent

```python
#Define Panel Widgets

year_slider = pn.widgets.IntSlider(name = 'Year slider',
                                   start=1750, 
                                   end = 2020, 
                                   step=5, 
                                   value=1850)
```

```python
year_slider
```

```python
#Radio buttons for CO2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name = "Y axis",
    options = ['co2', 'co2_per_capita',],
    button_type = "success"
)
```

```python
continents = ['World', 'Asia', 'Oceana', 'Europe', 
              'Africa', 'North America', 'South America', 
              'Antarctica']
idf = df.interactive()

co2_pipeline = (idf[(idf.year <= year_slider) &
                    (idf.country.isin(continents))]
                .groupby(['country','year'])[yaxis_co2].mean()
                .to_frame()
                .reset_index()
                .sort_values(by='year')
                .reset_index(drop = True)
                )

co2_pipeline
```

```python
#Radio buttions for CO2 measures
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name = 'Y axis',
    options = ['co2','co2_per_capita',],
    button_type = 'success'
)
```

```python
co2_plot = co2_pipeline.hvplot(x='year',
                               by = 'country',
                               y = yaxis_co2, 
                               line_width = 2,
                               title = 'CO2 emission by continent')
co2_plot
```

CO2 emmissions over time by continent
(Table style)

```python
co2_table = co2_pipeline.pipe(pn.widgets.Tabulator,
                             pagination = "remote",
                             page_size = 5,
                             sizing_mode ="stretch_width")
co2_table
```

C02 vs GDP scatterplot

```python
# Create another pipeline for scatterplot
co2_vs_gdp_scatterplot_pipeline = (
    idf[
        (idf.year == year_slider) &
        (~(idf.country.isin(continents)))
        ]
                .groupby(['country','year'])['co2'].mean()
                .to_frame()
                .reset_index()
                .sort_values(by='year')
                .reset_index(drop = True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

Bar chart with CO2 sources by continent

```python
yaxis_co2_sources = pn.widgets.RadioButtonGroup(
    name = 'Y axis',
    options = ['coal_co2', 'oil_co2', 'gas_co2'],
    button_type = 'success'
)

continents_excl_world = ['World', 'Asia', 'Oceana', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']
co2_source_bar_pipeline = (
    idf[
        (idf.year == year_slider) &
        (idf.country.isin(continents_excl_world))
        ]
    .groupby(['country','year'])[yaxis_co2_sources].sum()
                .to_frame()
                .reset_index()
                .sort_values(by='year')
                .reset_index(drop = True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind = 'bar',
                                                     x = 'country',
                                                     y = yaxis_co2_sources,
                                                     title = 'CO2 source by continent')
co2_source_bar_plot
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind = 'bar',
                                                     x = 'country',
                                                     y = yaxis_co2_sources,
                                                     title = 'CO2 source by continent')
co2_source_bar_plot
```

Dashboard

```python
#Layout using template from plan website
template = pn.template.FastListTemplate(
    title = 'World CO2 emission dashboard',
    sidebar = [pn.pane.Markdown("# CO2 Emissions and Climate Change"),
               pn.pane.Markdown('#### Carbon diaxode emissions are the primary delivery of the glibal climate change. '),
               pn.pane.PNG('climate_day.png', sizing_mode = 'scale_both'),
               pn.pane.Markdown("#### Settings"),
               year_slider],
    main =[pn.Row(pn.Column(yaxis_co2, co2_plot.panel(width=600, margin=(0,25))),
                  pn.Column(co2_table.panel(width =600, margin_left=(0,25)))),
           pn.Row(pn.Column(co2_vs_gdp_scatterplot.panel(width =600), margin=(0,25)),
                  pn.Column(yaxis_co2_sources, co2_source_bar_plot.panel(width = 600)))],
           accent_base_color = "#88d8b0",
           header_background ="#88d8b0",       
                  )
    
template.show()
template.servable();
```
